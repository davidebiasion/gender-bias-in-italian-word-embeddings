# Agency-Communion distribution

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
import statistics as stat

# mine
import db_functions as db
```

### All words

```python
# words
ag_com = db.read('ag_com_mf_ita.txt', 'ag-com-mf')

words = []
ag = []
com = []

for t in ag_com:
    words.append(t[0])
    ag.append(t[2])
    com.append(t[3])
```

```python
# agency communion correlation
r_all = scipy.stats.pearsonr(ag, com)[0]
p_all = scipy.stats.pearsonr(ag, com)[1]
print("r: %.3f" % (r_all))
print("p: %.3f" % (p_all))
```

```python
# distribution plot
ag.sort()
com.sort()

fig = plt.figure(figsize=(10,7))    
ax = fig.add_subplot(111)
ax.plot(np.arange(len(ag)), ag, linewidth=0, marker='s', color='blue', label='all words')
ax.set_ylabel('agency score')
ax.legend(facecolor='white')
ax.axes.get_xaxis().set_visible(False)


plt.show()
fig.savefig('distr-plot/ag-distr-all.pdf', bbox_inches='tight', format='pdf')
```

```python
fig = plt.figure(figsize=(10,7)) 
ax = fig.add_subplot(111)
ax.plot(np.arange(len(com)), com, linewidth=0, marker='s', color='pink', label='all words')
ax.set_ylabel('communion score')
ax.legend(facecolor='white')
ax.axes.get_xaxis().set_visible(False)

plt.show()
fig.savefig('distr-plot/com-distr-all.pdf', bbox_inches='tight', format='pdf')
```

```python
# statistics
ag_mean = stat.mean(ag)
ag_median = stat.median(ag)
ag_stdev = stat.stdev(ag)

com_mean = stat.mean(com)
com_median = stat.median(com)
com_stdev = stat.stdev(com)
```

```python
print("ag_mean:   %2.3f" % (ag_mean))
print("ag_median: %2.3f" % (ag_median))
print("ag_stdev:  %2.3f" % (ag_stdev))
```

```python
print("com_mean:   %2.3f" % (com_mean))
print("com_median: %2.3f" % (com_median))
print("com_stdev:  %2.3f" % (com_stdev))
```

### Ungendered words

```python
# words
ag_com_ung = db.read('ag_com_ung_ita.txt', 'ag-com-mf')

words_ung = []
ag_ung = []
com_ung = []

for t in ag_com_ung:
    words_ung.append(t[0])
    ag_ung.append(t[2])
    com_ung.append(t[3])
```

```python
# agency communion correlation
r_ung = scipy.stats.pearsonr(ag_ung, com_ung)[0]
p_ung = scipy.stats.pearsonr(ag_ung, com_ung)[1]
print("r: %.3f" % (r_ung))
print("p: %.3f" % (p_ung))
```

```python
# bar chart
x = np.arange(len(words_ung))  # the label locations
width = 0.2  # the width of the bars

fig = plt.figure(figsize=(20,7))    
ax = fig.add_subplot(111)
rects1 = ax.bar(x - width/2, ag_ung, width, color='blue', label='agency')
rects2 = ax.bar(x + width/2, com_ung, width, color='pink', label='communion')
ax.set_ylabel('score')
ax.axes.get_xaxis().set_visible(False)
ax.legend()


for i, label in enumerate(words_ung):
    plt.annotate(label, # this is the text
                 (i, -1.5), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,0), # distance from text to points (x,y)
                 ha='center', # horizontal alignment can be left, right or center
                 rotation=90)

plt.show()
fig.savefig('distr-plot/bar-ung.pdf', bbox_inches='tight', format='pdf')
```

```python
# distribution plot
ag_ung.sort()
com_ung.sort()

fig = plt.figure(figsize=(20,7))    
ax = fig.add_subplot(121)
ax.plot(np.arange(len(ag_ung)), ag_ung, linewidth=0, marker='s', color='blue', label='Agency')
ax.set_ylabel('agency score')
ax.set_title("Ungendered words")
ax.legend(facecolor='white')
ax.axes.get_xaxis().set_visible(False)

ax = fig.add_subplot(122)
ax.plot(np.arange(len(com_ung)), com_ung, linewidth=0, marker='s', color='pink', label='Communion')
ax.set_ylabel('communion score')
ax.set_title("Ungendered words")
ax.legend(facecolor='white')

plt.show()
```

```python
# statistics
ag_ung_mean = stat.mean(ag_ung)
ag_ung_median = stat.median(ag_ung)
ag_ung_stdev = stat.stdev(ag_ung)

com_ung_mean = stat.mean(com_ung)
com_ung_median = stat.median(com_ung)
com_ung_stdev = stat.stdev(com_ung)
```

```python
print("ag_ung_mean:    %2.3f" % (ag_ung_mean))
print("ag_ung_median:  %2.3f" % (ag_ung_median))
print("ag_ung_stdev:   %2.3f" % (ag_ung_stdev))
```

```python
print("com_ung_mean:   %2.3f" % (com_ung_mean))
print("com_ung_median: %2.3f" % (com_ung_median))
print("com_ung_stdev:  %2.3f" % (com_ung_stdev))
```
