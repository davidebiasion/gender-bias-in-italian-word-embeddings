```python
# gender direction
g_pca_0  = pc[0]/LA.norm(pc[0])
g_pca_01 = (sv[0]*pc[0]+sv[1]*pc[1])/(sv[0]+sv[1])/LA.norm((sv[0]*pc[0]+sv[1]*pc[1])/(sv[0]+sv[1]))
```

#### Step 2.3: pseudo-LDA

```python
'''
S_word = [['lei','lui'],
          ['donna','uomo'],
          ['madre','padre'],
          ['moglie','marito'],
          ['sorella','fratello'],
          ['femmina','maschio']]
'''

S_word = [['lui','lei'],
          ['uomo','donna'], 
          ['padre','madre'],
          ['marito','moglie'],
          ['fratello','sorella'],
          ['maschio','femmina']]

G_word = db.read('gram_def_mf.txt', 'mf')
```

```python
S_m_v = []
for w in S_word:
    S_m_v.append(E_ft.v(w[1]))
    
S_f_v = []
for w in S_word:
    S_f_v.append(E_ft.v(w[0]))
    
G_m_v = []
for w in G_word:
    G_m_v.append(E_ft.v(w[0]))
    
G_f_v = []
for w in G_word:
    G_f_v.append(E_ft.v(w[1]))

S_m = np.array(S_m_v).T
S_f = np.array(S_f_v).T
G_m = np.array(G_m_v).T
G_f = np.array(G_f_v).T
```

```python
# delta semantic
Delta_S = np.dot(np.subtract(S_m, S_f), np.subtract(S_m, S_f).T)

# delta grammatical
Delta_G = np.dot(np.subtract(G_m, G_f), np.subtract(G_m, G_f).T)

# A
A = np.dot(np.linalg.inv(Delta_G), Delta_S)
```

```python
# eigenvalues and eigenvectors
w, v = np.linalg.eig(A)

# gender direction: pick the eigenvector associated to the largest eigenvalue
g_e = v[:,np.argmax(w)] # already normalized
```

```python
# gendered professions subsets
gen_group_0 = [] # 0%  <= %female < 20%
gen_group_1 = [] # 20% <= %female < 40%
gen_group_2 = [] # 40% <= %female < 60%
gen_group_3 = [] # 60% <= %female < 80%
gen_group_4 = [] # 80% <= %female <= 100%

for i in range(len(gen_col0)):
    score = gen_col10[i]
    item = [gen_col0[i], 
            gen_col1[i]-gen_col6[i],
            gen_col2[i]-gen_col7[i],
            gen_col3[i]-gen_col8[i],        
            gen_col4[i]-gen_col9[i],            
            score]
    if score < 20:
        gen_group_0.append(item)
    
    elif score >= 20 and score < 40:
        gen_group_1.append(item)
        
    elif score >= 40 and score < 60:
        gen_group_2.append(item)

    elif score >= 60 and score < 80:
        gen_group_3.append(item)

    elif score >= 80:
        gen_group_4.append(item)
```

```python
def compute_avg(group):
    avg_g_diff = 0
    avg_g_pca_0 = 0
    avg_g_pca_01 = 0
    avg_g_e = 0
    for item in group:
        avg_g_diff = avg_g_diff + item[1]/len(group)
        avg_g_pca_0 = avg_g_pca_0 + item[2]/len(group)
        avg_g_pca_01 = avg_g_pca_01 + item[3]/len(group)
        avg_g_e = avg_g_e + item[4]/len(group)
        
    return [avg_g_diff, avg_g_pca_0, avg_g_pca_01, avg_g_e]
```

```python
# subgroup average differences of gendered projections
group_0_ext = compute_avg(gen_group_0)
group_1_ext = compute_avg(gen_group_1)
group_2_ext = compute_avg(gen_group_2)
group_3_ext = compute_avg(gen_group_3)
group_4_ext = compute_avg(gen_group_4)
```

```python
S_m_v = []
for w in S_word:
    S_m_v.append(E_ft.v(w[1]))
    
S_f_v = []
for w in S_word:
    S_f_v.append(E_ft.v(w[0]))
    
G_m_v = []
for w in G_word:
    G_m_v.append(E_ft.v(w[0]))
    
G_f_v = []
for w in G_word:
    G_f_v.append(E_ft.v(w[1]))

S_m = np.array(S_m_v).T
S_f = np.array(S_f_v).T
G_m = np.array(G_m_v).T
G_f = np.array(G_f_v).T
```

```python
# delta semantic
Delta_S = np.dot(np.subtract(S_m, S_f), np.subtract(S_m, S_f).T)

# delta grammatical
Delta_G = np.dot(np.subtract(G_m, G_f), np.subtract(G_m, G_f).T)

# A
A = np.dot(np.linalg.inv(Delta_G), Delta_S)
```

```python
# eigenvalues and eigenvectors
w, v = np.linalg.eig(A)

# gender direction: pick the eigenvector associated to the largest eigenvalue
g_e = v[:,np.argmax(w)] # already normalized
```

```python
'''
S_word = [['lei','lui'],
          ['donna','uomo'],
          ['madre','padre'],
          ['moglie','marito'],
          ['sorella','fratello'],
          ['femmina','maschio']]
'''

S_word = [['lui','lei'],
          ['uomo','donna'], 
          ['padre','madre'],
          ['marito','moglie'],
          ['fratello','sorella'],
          ['maschio','femmina']]

G_word = db.read('gram_def_mf.txt', 'mf')
```

```python
S_m_v = []
for w in S_word:
    S_m_v.append(E_g.v(w[1]))
    
S_f_v = []
for w in S_word:
    S_f_v.append(E_g.v(w[0]))
    
G_m_v = []
for w in G_word:
    G_m_v.append(E_g.v(w[0]))
    
G_f_v = []
for w in G_word:
    G_f_v.append(E_g.v(w[1]))

S_m = np.array(S_m_v).T
S_f = np.array(S_f_v).T
G_m = np.array(G_m_v).T
G_f = np.array(G_f_v).T
```

```python
# delta semantic
Delta_S = np.dot(np.subtract(S_m, S_f), np.subtract(S_m, S_f).T)

# delta grammatical
Delta_G = np.dot(np.subtract(G_m, G_f), np.subtract(G_m, G_f).T)

# A
A = np.dot(np.linalg.inv(Delta_G), Delta_S)
```

```python
# eigenvalues and eigenvectors
w, v = np.linalg.eig(A)

# gender direction: pick the eigenvector associated to the largest eigenvalue
g_e = v[:,np.argmax(w)] # already normalized
```
