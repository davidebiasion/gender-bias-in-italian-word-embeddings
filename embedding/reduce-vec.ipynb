```python
# read list of words
file = open('reduced-list-en.txt', 'r')
lines = file.readlines()
words_list = []
for line in lines:
    words_list.append(line[:-1])
file.close()

print("Words list of "+str(len(words_list))+" words")
```

```python
# extract only the embeddings we want
file_name = 'fast-text/wiki.en.reduced.vec'
file = open(file_name, 'w')
vec_list = []
inserted_list = []
with open('fast-text/wiki.en.vec', "r", encoding='utf8') as f:
    i = 0
    for line in f:
        s = line.split()
        if s[0] in words_list and s[0] not in vec_list:
            file.write(line)
            vec_list.append(s[0])
            print("word %4d: %s" % (i,s[0]))
            i = i+1

    print("*** writing completed: "+file_name+" of "+str(i)+" words")

file.close()
```

```python
# check missing words
missing_list = []
i = 0
for word in words_list:
    if word not in vec_list:
        i = i+1
        missing_list.append(word)

print("Words not inserted: "+str(missing_list))
```
